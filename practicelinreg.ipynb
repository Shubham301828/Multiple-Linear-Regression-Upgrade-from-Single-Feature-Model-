```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = {'Hours studied': [1, 2, 3, 4, 5], 'Hours Slept': [5, 6, 6, 7, 8], 'Marks obtained': [30, 45, 50, 65, 80]}
df = pd.DataFrame(data)
x = df[['Hours studied', 'Hours Slept']]
y = df['Marks obtained']
model = LinearRegression()
model.fit(x, y)
pred_marks = model.predict([[6,7]])
print(f"Predicted marks for 6 hours studied and 7 hours slept: {pred_marks[0]:.2f}")
y_pred = model.predict(x)
plt.scatter(df['Hours studied'], y, color='blue', label='Actual')
plt.plot(df['Hours studied'], y_pred, color='red', label='Prediction Line')
plt.xlabel('Hours studied')
plt.ylabel('Marks')
plt.legend()
plt.show()
```
